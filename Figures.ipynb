```python
import pickle
import numpy as np

def visualize_from_pickle(probabilities_pickle_files, num_copies_lists, output_filenames):
    # Check that all input lists have the same length
    assert len(probabilities_pickle_files) == len(num_copies_lists) == len(output_filenames)

    fig, axes = plt.subplots(nrows=1, ncols=len(probabilities_pickle_files),figsize=(12, 5), sharey=True)

    for idx, (pickle_file, num_copies_list, output_filename) in enumerate(zip(probabilities_pickle_files, num_copies_lists, output_filenames)):
        # Load the word probabilities from the pickle file
        with open(pickle_file, "rb") as f:
            word_probabilities = pickle.load(f)

        a_b = ["(a) GPT Neo 125M", "(b) GPT Neo 350M"]
        # Plot the data for non-empty lists
        for i, word_probs in enumerate(word_probabilities):
            if not word_probs:  # Skip empty lists
                continue
            x = list(range(1, len(word_probs) + 1))
            y = [p for _, p in word_probs]

            # Compute rolling mean of y-values
            window = 20
            weights = np.repeat(1.0, window) / window
            y_smooth = np.convolve(y, weights, 'valid')

            # Adjust x-values to match smoothed y-values
            x_smooth = x[window // 2:-(window // 2) or None]

            # Plot the smoothed line
            axes[idx].plot(x_smooth, y_smooth[1:], label=f"Num Copies: {num_copies_list[i]}", color=f"C{i}", linewidth=0.8)

        # Configure the plot
        if idx == 0:
            axes[idx].set_ylabel("Token probability")
        axes[idx].set_xlabel(f"Token position\n\n{a_b[idx]}")
        axes[idx].legend(loc='lower right')

        # Save the plot to a file
        plt.savefig(output_filename,bbox_inches='tight', dpi=600)

    # Close the plot to free up memory
    plt.close(fig)


# Usage example
visualize_from_pickle(["/Users/luka.borec/gpt-neo-125M-2023-03-03-11h00m00s_word_probabilities.pkl", "/Users/luka.borec/gpt-neo-350M-2023-03-07-19h11m23s_word_probabilities.pkl"], 
                      [[1, 10, 20, 30], [1, 10, 20, 30]], 
                      ["visualized_probabilities_125M.png", "visualized_probabilities_350M.png"])
```

```python
import json
from difflib import SequenceMatcher
import pdfkit
path_wkhtmltopdf = r'/usr/local/bin/wkhtmltopdf'
config = pdfkit.configuration(wkhtmltopdf=path_wkhtmltopdf)
pdfkit.from_file('report.html', 'report.pdf', configuration=config)


def compare_texts(original, decoded):
    seq_matcher = SequenceMatcher(None, original.split(), decoded.split())
    original_words = []
    decoded_words = []
    
    for opcode, a0, a1, b0, b1 in seq_matcher.get_opcodes():
        if opcode == "equal":
            original_words.extend([f'<span style="color:green">{word}</span>' for word in original.split()[a0:a1]])
            decoded_words.extend([f'<span style="color:green">{word}</span>' for word in decoded.split()[b0:b1]])
        else:
            original_words.extend([f'<span style="color:red">{word}</span>' for word in original.split()[a0:a1]])
            decoded_words.extend([f'<span style="color:red">{word}</span>' for word in decoded.split()[b0:b1]])
    
    return ' '.join(original_words), ' '.join(decoded_words)

with open('/Users/lukaborec/gpt-neo-125M-2023-03-03-11h00m00s_memorization_coherence_results.json', 'r') as file:
    data = json.load(file)

html_report = "<html><body>"

for structure in data:
    if not structure['memorized']:
        num_copies = structure['num_copies']
        original, decoded = compare_texts(structure['original'], structure['decoded'])
        html_report += f'<p>Num copies:<br>{num_copies}</p><p>Original:<br>{original}</p><p>Decoded:<br>{decoded}</p>'

html_report += "</body></html>"

# Write the HTML report to a file
with open("report.html", "w") as f:
    f.write(html_report)

# Convert the HTML report to a PDF
pdfkit.from_file('report.html', 'report.pdf')
```

```python
def detect_consecutive_repetition(text):
    # Preprocess the text to remove punctuation and make it lower case
    text = re.sub(r'[^\w\s]', '', text.lower())

    # Split the text into a list of words
    words = text.split()

    # Initialize a dictionary to store the results
    repeated_phrases = defaultdict(int)

    # Loop over the sizes of phrases
    for n in range(3, 10):
        # Generate the phrases for this size
        phrases = list(ngrams(words, n))
        # Convert the phrases to strings
        phrases_str = [' '.join(phrase) for phrase in phrases]

        # Check each phrase
        count = defaultdict(int)
        prev_index = defaultdict(int)
        for i in range(len(phrases_str)):
            # The phrase is at phrases_str[i]
            phrase = phrases_str[i]
            
            # Skip undesired phrases
            if "pad pad pad" in phrase:
                continue 

            # Check if the phrase is repeated consecutively or with a gap of n
            if i - prev_index[phrase] == n:
                count[phrase] += 1
            else:
                if count[phrase] >= 3:
                    repeated_phrases[phrase] = count[phrase]
                count[phrase] = 1
            
            prev_index[phrase] = i

        for phrase, c in count.items():
            if c >= 3:
                repeated_phrases[phrase] = c

    return repeated_phrases
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.pdfgen import canvas

def highlight_and_save(text_keyword_pairs, filename):
    c = canvas.Canvas(filename, pagesize=letter)
    width, height = letter
    margin = 10
    max_x = width - margin
    max_y = height - margin
    font_size = 8
    c.setFont("Courier", font_size)
    line_height = font_size * 1.5
    max_line_width = max_x - margin

    total_height = 0
    for text, _ in text_keyword_pairs:
        lines = len(text.split()) // (max_line_width // (font_size // 2)) + 1
        total_height += lines * line_height + line_height * 2

    y_position = max_y - (max_y - total_height) / 2

    for text, keyword in text_keyword_pairs:
        x_position = margin
        words = text.split()
        line = ""
        
        for word in words:
            temp_line = line + word + " "
            if c.stringWidth(temp_line, 'Courier', font_size) < max_line_width:
                line = temp_line
            else:
                draw_highlighted_line(c, line, keyword, x_position, y_position, font_size)
                y_position -= line_height
                line = word + " "

        if line.strip() != "":
            draw_highlighted_line(c, line, keyword, x_position, y_position, font_size)

        y_position -= line_height * 2  # Extra space between texts

    c.save()

def draw_highlighted_line(c, line, keyword, x_position, y_position, font_size):
    index = 0
    while index < len(line):
        keyword_index = line.lower().find(keyword.lower(), index)
        if keyword_index == -1:
            c.drawString(x_position, y_position, line[index:])
            break

        c.drawString(x_position, y_position, line[index:keyword_index])
        x_position += c.stringWidth(line[index:keyword_index], 'Courier', font_size)
        c.setFillColor(colors.yellow)
        c.rect(x_position, y_position - 2, c.stringWidth(keyword, 'Courier', font_size), font_size + 4, fill=1, stroke=0)
        c.setFillColor(colors.black)
        c.drawString(x_position, y_position, line[keyword_index:keyword_index + len(keyword)])
        x_position += c.stringWidth(keyword, 'Courier', font_size)
        index = keyword_index + len(keyword)
```

```python
from PyPDF2 import PdfReader, PdfWriter, PdfMerger

def crop_pdf(input_pdf, output_pdf):
    pdf_reader = PdfReader(input_pdf)
    pdf_writer = PdfWriter()

    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        crop_box = page.mediabox.lower_left, page.mediabox.upper_right

        # Crop the page to the text bounding box
        page.mediabox.lower_left = crop_box[0]
        page.mediabox.upper_right = crop_box[1]
        pdf_writer.add_page(page)

    with open(output_pdf, "wb") as f:
        pdf_writer.write(f)

def merge_and_crop_pdfs(pdf_list, output_filename):
    merger = PdfMerger()

    for pdf in pdf_list:
        cropped_pdf = f"cropped_{pdf}"
        crop_pdf(pdf, cropped_pdf)
        merger.append(cropped_pdf)

    merger.write(output_filename)
    merger.close()


pdf_list = ["repetition___vaccine is not live, and the.pdf", 
            "repetition___run_python_file exec_code_object(code, main_mod.__dict__) File exec_function> line.pdf", 
            "repetition___They’re a pretty good team.pdf"]
output_filename = "merged_cropped.pdf"
merge_and_crop_pdfs(pdf_list, output_filename)
```

```python
import json
from nltk.translate.bleu_score import sentence_bleu

# List of file names
files = ['/Users/luka.borec/Downloads/sth/gpt-neo-125M-2023-03-03-11h00m00s_nucleus_sampling_0.2.json',
 '/Users/luka.borec/Downloads/sth/gpt-neo-125M-2023-03-03-11h00m00s_nucleus_sampling_0.4.json',
 '/Users/luka.borec/Downloads/sth/gpt-neo-125M-2023-03-03-11h00m00s_nucleus_sampling_0.6.json',
 '/Users/luka.borec/Downloads/sth/gpt-neo-125M-2023-03-03-11h00m00s_nucleus_sampling_0.8.json',
 '/Users/luka.borec/Downloads/sth/gpt-neo-350M-2023-03-07-19h11m23s_nucleus_sampling_0.2.json',
 '/Users/luka.borec/Downloads/sth/gpt-neo-350M-2023-03-07-19h11m23s_nucleus_sampling_0.4.json',
 '/Users/luka.borec/Downloads/sth/gpt-neo-350M-2023-03-07-19h11m23s_nucleus_sampling_0.6.json',
 '/Users/luka.borec/Downloads/sth/gpt-neo-350M-2023-03-07-19h11m23s_nucleus_sampling_0.8.json']


results = {}

for file in files:
    # Extract model name and top_k parameter from file name
    model_name = file.split('-')[2]
    top_k = float(file.split('_')[-1].replace('.json', ''))
    
    # Initialize dictionary for model if not already present
    if model_name not in results:
        results[model_name] = {}
    if top_k not in results[model_name]:
        results[model_name][top_k] = {}
    
    # Read file content
    with open(file, 'r') as f:
        data = json.load(f)
    
    # Compute BLEU scores and separate by "num_copies"
    for item in data:
        if item["memorized"]:
            continue
        num_copies = item["num_copies"]
#         print(len(item['original'].split()))
#         print(len(item['decoded'].split()))
#         print()
        half = len(item['original'].split()) //2
        reference = item['original'].split()[half:]
        candidate = item['decoded'].split()[half:]
        score = sentence_bleu([reference], candidate, weights=(0.25, 0.25, 0.25, 0.25))
        
        if num_copies not in results[model_name][top_k]:
            results[model_name][top_k][num_copies] = {'total_score': 0, 'count': 0}
        
        results[model_name][top_k][num_copies]['total_score'] += score
        results[model_name][top_k][num_copies]['count'] += 1

# Compute average BLEU scores
for model in results:
    for top_k in results[model]:
        for num_copies in results[model][top_k]:
            total_score = results[model][top_k][num_copies]['total_score']
            count = results[model][top_k][num_copies]['count']
            avg_score = total_score / count
            results[model][top_k][num_copies] = avg_score
```

```python
import os
import re
import pickle
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns

def calculate_entropy(pickle_dir):
    top_p_values = [0.8, 0.6, 0.4, 0.2]  # Reversed for decreasing order
    model_sizes = ['125M', '350M']  # 125M on left, 350M on right
    num_copies_list = [1, 5, 15, 25]

    # Dataframe to store results
    data = {
        'model_size': [],
        'top_p': [],
        'num_copies': [],
        'average_entropy': []
    }

    for i, top_p in enumerate(top_p_values):
        for j, model_size in enumerate(model_sizes):
            # Find corresponding pickle file
            pattern = f"gpt-neo-{model_size}.*_word_probabilities_{top_p}.pkl"
            for k, fname in enumerate(sorted(os.listdir(pickle_dir))):
                if re.fullmatch(pattern, fname):
                    with open(os.path.join(pickle_dir, fname), 'rb') as f:
                        word_probabilities = pickle.load(f)

                    total_entropy = 0
                    valid_sentences = 0
                    for word_probs in word_probabilities:
                        if not word_probs:  # Skip empty lists
                            continue
                        # Calculate entropy
                        entropy = -sum(p * np.log2(p) for p in word_probs if p > 0)  # Exclude zero probabilities
                        total_entropy += entropy
                        valid_sentences += 1

                    if valid_sentences > 0:
                        average_entropy = total_entropy / valid_sentences
                        data['model_size'].append(model_size)
                        data['top_p'].append(top_p)
                        data['num_copies'].append(num_copies_list[k])
                        data['average_entropy'].append(average_entropy)

    return pd.DataFrame(data)

# Call the function
df = calculate_entropy("/Users/luka.borec/Downloads/Archive")

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='top_p', y='average_entropy', hue='num_copies', marker='o', palette='viridis')

plt.title('Effect of top_p on Average Entropy for Different Number of Copies')
plt.xlabel('top_p Value')
plt.ylabel('Average Entropy')
plt.legend(title='Number of Copies', loc='upper right')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
