```python
import os
from datasets import load_dataset

root_dir = "memorization/dataset/sampled_dataset"

file_list = []

for subdir, _, files in os.walk(root_dir):
    for file in files:
        if file.endswith('.txt'):
            file_list.append(os.path.join(subdir, file))

dataset = load_dataset("text", data_dir="memorization/dataset/sampled_dataset", sample_by="document")
```

```python
train_dataset = dataset
from transformers import Trainer, TrainingArguments, GPT2LMHeadModel, GPT2Tokenizer, AutoConfig, AutoTokenizer, \
    DataCollatorForLanguageModeling
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.add_special_tokens({'pad_token': '[PAD]'})
```

```python
CONTEXT_LENGTH=512
```

```python
def tokenize(element):
    text = "<|endoftext|> " + element["text"] + " <|endoftext|>"
    outputs = tokenizer(
        text,
        truncation=True,
        max_length=CONTEXT_LENGTH,
    )
    outputs['input_ids'][-1] = tokenizer.eos_token_id
    return {"input_ids": outputs['input_ids']}
```

```python
train_dataset_tokenized = train_dataset.map(tokenize, batched=False)
```
